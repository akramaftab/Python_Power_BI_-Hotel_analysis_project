# Data Analysis and Visualization

```python
df["is_canceled"].value_counts(normalize = True) * 100
```

```python
def can(value):
    if value == 0:
        return "Not cancelled"
    else:
        return "Cancelled"
df["is_canceled"] = df["is_canceled"].apply(can)
```

```python
plt.figure(figsize=(4,4))
gb = df.groupby("is_canceled").agg({'is_canceled':"count"})
plt.pie(gb["is_canceled"],labels = gb.index,autopct = "%1.2f%%")
plt.title("Percentage of Reservations Status",fontsize = 10)
plt.show()
```

 ##About 63% of hotel reservations were not canceled, while roughly 37% ended up being canceled.

```python
plt.figure(figsize=(7,5))
ax = sns.countplot( x = "hotel", hue = "is_canceled", data = df)
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[0])
plt.ylabel("Number of Reservations")
plt.title("Reservations Status by hotels", fontsize = 10)
plt.legend(title="Reservation Status", bbox_to_anchor = (1,1))
plt.show()
```

##City hotels had both higher total bookings and a much higher cancellation count compared to resort hotels.

```python
resort_hotel = df[df["hotel"] == "Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize = True)*100
```

```python
city_hotel = df[df["hotel"] == "City Hotel"]
city_hotel["is_canceled"].value_counts(normalize = True)*100
```

```python
resort_hotel = resort_hotel.groupby("arrival_date")[["adr"]].mean()
city_hotel = city_hotel.groupby("arrival_date")[["adr"]].mean()
```

```python
plt.figure(figsize = (15,4))
plt.title("Average dailry rate in City and Resort Hotel", fontsize = 15)
plt.plot(resort_hotel.index,resort_hotel["adr"], label = "Resort Hotel")
plt.plot(city_hotel.index,city_hotel["adr"], label = "City Hotel")
plt.legend(title="Average Daily Rate", fontsize = 10)
plt.show()
```

##The average daily rate (price) for city hotels was consistently higher and more volatile than for resort hotels, especially during peak periods.

```python
df["month"] = df["arrival_date"].dt.month
plt.figure(figsize = (12,6))
ax = sns.countplot(x = "month", hue = "is_canceled", data = df)
ax.legend(title="Reservation Status", bbox_to_anchor = (1,1), fontsize = 8)
plt.title("Reservation Status by Month", fontsize = 10)
plt.xlabel("month")
plt.ylabel("No. of reservations")
plt.show()
```

##Most reservations and cancellations happened between March and August, with both dropping noticeably in the winter months like November, December, and January.

```python
canceled_reservations = df[df["is_canceled"] == "Cancelled"]
monthly_adr = canceled_reservations.groupby("month")[["adr"]].sum().reset_index()
plt.figure(figsize=(9,4))
plt.title("ADR per month")
sns.barplot(x="month", y="adr", data=monthly_adr)
plt.show()
```

##The total monthly ADR from canceled reservations peaked in the summer, especially August, indicating higher booking prices that likely influenced cancellation rates during busy seasons.

```python
plt.figure(figsize = (9,4))
plt.title("Reservation cancellations by waiting list")
sns.histplot(x="days_in_waiting_list", data = canceled_reservations ,bins = 100)
plt.show()
```

##Almost all cancellations happened at zero days on the waiting list, showing guests who book last-minute are much more likely to cancel.

```python
def gan(value):
    if value == 0:
        return "New Guests"
    else:
        return "Repeated Guest"
df["is_repeated_guest"] = df["is_repeated_guest"].apply(gan)
```

```python
df["is_repeated_guest"].value_counts(normalize = True) * 100
```

```python
plt.figure(figsize=(7,5))
ax = sns.countplot( x = "is_repeated_guest", hue = "is_canceled", data = df)
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[0])
plt.ylabel("Number of Reservations")
plt.xlabel("Guest Type")
plt.title("Reservations Status by Guest type", fontsize = 10)
plt.legend(title="Reservation Status", bbox_to_anchor = (1,1))
plt.show()
```

##Cancellations were far more common among new guests than repeat guests; new customers are much more likely to cancel their bookings.
